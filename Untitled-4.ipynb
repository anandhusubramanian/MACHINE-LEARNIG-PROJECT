```python
x = df[['AnnualIncome','Score']]
x.head()
```

##### DETERMINING THE NUMBER OF K-CLUSTERS

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,19):
    m = KMeans(n_clusters=i , max_iter=300)
    m.fit(x)  
    wcss.append(m.inertia_)
plt.plot(list(range(1,19)),wcss)
plt.show()
```

##### THE CURVE IS FLATTENED AT 5.0 SO WE WILL CHOOSE 5 AS THE K-CLUSTER

##### DATA MODELLING

```python
Kmodel = KMeans(n_clusters = 5)
Kmodel.fit(x)
```

```python
y_pred = Kmodel.predict(x)
print("All the cluster values:",set(y_pred))
```

```python
centroid = Kmodel.cluster_centers_
centroid
```

##### THIS ARE THE FIVE CENTROIDS IN ANNUALINCOME VS SCORE PLOT.LETS MAKE A STRIPPLOT OUT OF IT

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=y_pred)
plt.scatter(centroid[:,0],centroid[:,1],marker='*',color='red',s=150)
plt.xlabel('AnnualIncome')
plt.ylabel('Score')
plt.title('K-Clusters')
plt.show()
```

##### THIS ARE OUR FINAL CLUSTERS AND CENTROIDS
